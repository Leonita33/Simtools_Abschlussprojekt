This notebook evaluates stored simulation results.

- **Question 1:** Which parameters have the greatest influence on the risk of over-indebtedness?

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

**Question 1:** Which parameters have the greatest influence on the risk of over-indebtedness?

Data source: data/results/factors_influencing_over-indebtedness.csv

```python
df_imp = pd.read_csv("../data/results/factors_influencing_over-indebtedness.csv")
df_imp
```

```python
plt.figure(figsize=(7,4))
plt.bar(df_imp["parameter"], df_imp["importance"])
plt.title("Strength of influence parameters on the risk of over-indebtedness")
plt.xlabel("parameters")
plt.ylabel("strength of influence")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()
```

### Interpretation (question 1)
- **Income** and **addiction_strength** have the greatest influence on the risk of over-indebtedness.
- **Savings_buffer** acts as a buffer.
- **Interest_rate** is relevant, but less dominant in comparison. 

**Question 2:** How have key indicators (compulsive buying, debt) changed between 1995 and 2025?

Data source: data/results/time_series_key_figures_1995_2025.csv

```python
df_ts = pd.read_csv("../data/results/time_series_key_figures_1995_2025.csv")
df_ts["year"] = df_ts["year"].astype(int)
df_ts.head()
df_ts[df_ts["year"].isin([1995, 2005, 2015, 2025])]
```

```python
plt.figure(figsize=(8,5))
plt.plot(df_ts["year"], df_ts["addiction_rate"], marker="o")
plt.title("Development of the purchase rate (1995-2025)")
plt.xlabel("year")
plt.ylabel("Purchase search rate (%)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df_ts["year"], df_ts["mean_debt"], marker="o")
plt.title("Development of average debt (1995-2025)")
plt.xlabel("year")
plt.ylabel("average debt (â‚¬)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df_ts["year"], df_ts["pct_over_50"], marker="o")
plt.title("Proportion with debt > 50% of income (1995-2025)")
plt.xlabel("year")
plt.ylabel("percentage (%)")
plt.grid(True)
plt.show()
```

### Interpretation (question 2)
- The search rate increases significantly over the period.
- Average debt also increases sharply.
- The proportion of heavily indebted individuals (debt > 50% of income) rises sharply.
